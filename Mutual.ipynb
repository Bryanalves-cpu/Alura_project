```python
#Importando Bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import numpy as np


```

```python
#Primeira conexão com banco de dados fornecido

Dados_Empresas = pd.read_csv("/content/empresas.csv")
Dados_Empresas
```

```python
Dados_Empresas.columns
```

```python
#Media de idade de empresa equivale para todos processos judiciais , 
#que podemos utilizar como um fator importante para determinar risco com 4 processos judiciais

Dados_Empresas.groupby('processos_judiciais_empresa')['idade_empresa'].mean()
     
```

```python
#Variavel importante para determinar risco , 
#fase eliminátoria deixaria como regra de recusa para empresas com 4 processos 
#1151 empresas com 4 processos judiciais existentes agora vamos analisar todos os fatores


Dados_Empresas['processos_judiciais_empresa'].value_counts()
```

```python
#Filtrando empresas com processos judiciais apartir de 4

Dados_processos = Dados_Empresas.query('processos_judiciais_empresa == 4')
Dados_processos
```

```python
#fator recusa 100% , empresas abaixo de 10 anos de empreendimento com 4 processos judiciais
# 496 empresas de 1151 , as restantes com mais de 10 anos de empreendimento teria uma chance maior de entrada

Fator_recusa4 = Dados_processos.query('idade_empresa < 10')
Fator_recusa4.count()
```
